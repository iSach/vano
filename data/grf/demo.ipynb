```python
import matplotlib.pyplot as plt
from grf import gen_grfs

cmap = plt.get_cmap('viridis')
nb = 32
for i in range(1, nb + 1, 4):
    X, u = gen_grfs(
        nb_fcts=1,
        alpha=2,
        tau=1,
        trunc_sum_elems=i,
        a=0,
        b=1,
        steps=128
    )


    #print(f"X: {X.shape}")
    #print(f"u: {u.shape}")

    plt.plot(X.T, u.T, c=cmap(int(i * 256 / nb)))

# Colorbar with the number of terms
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=0, vmax=nb))
sm._A = []
plt.colorbar(sm, label="Number of terms", ax=plt.gca())
plt.xlabel("X")
plt.ylabel("u")
plt.show()
```

```python
import matplotlib.pyplot as plt
from grf import gen_grfs
import numpy as np

X, u = gen_grfs(
    nb_fcts=256,
    alpha=2,
    tau=1,
    trunc_sum_elems=32,
    a=0,
    b=1,
    steps=128
)

plt.plot(X.T, u.T, c='k', alpha=0.1)
amp = u.abs().max()
plt.plot(X[0], amp * np.sin(2 * np.pi * X[0]), c='r', alpha=1.0)
plt.plot(X[0], -amp * np.sin(2 * np.pi * X[0]), c='r', alpha=1.0)
plt.show()
plt.plot(X.T, u.T)
plt.show()
```
